```python
from sklearn.linear_model import LogisticRegression
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib as plt
import statsmodels.api as sm
from patsy import dmatrices, dmatrix
```

```python
iris = pd.read_csv('data/Iris.data')
```

# Iris setosa

```python
iris_setosa = pd.DataFrame()

for key in iris.keys()[:-1]:
    iris_setosa[key] = iris[key]Âºa

iris_setosa['is_setosa'] = [1 if species == 'Iris-setosa' else 0 for species in iris['species']]
```

```python
sns.countplot(x='is_setosa', data=iris_setosa)
```

```python
sns.scatterplot(x="sepal_length",y="sepal_width", hue ="is_setosa", data = iris_setosa, palette='hls')
```

```python
sns.scatterplot(x="petal_length",y="petal_width", hue ="is_setosa", data = iris_setosa, palette='hls')
```

```python
iris_setosa.groupby('is_setosa').mean()
```

```python
pd.crosstab(iris_setosa.sepal_length, iris_setosa.is_setosa).plot(kind='bar', stacked=True)
```

```python
pd.crosstab(iris_setosa.petal_length, iris_setosa.is_setosa).plot(kind='bar', stacked=True)
```

```python
pd.crosstab(iris_setosa.sepal_width, iris_setosa.is_setosa).plot(kind='bar', stacked=True)
```

```python
pd.crosstab(iris_setosa.petal_width, iris_setosa.is_setosa).plot(kind='bar', stacked=True)
```

```python
from patsy import dmatrices
y,X = dmatrices('is_setosa ~ sepal_length + sepal_width', data = iris_setosa, return_type = 'dataframe')

logit_model = sm.Logit(y,X)
result = logit_model.fit()
print(result.summary2())
```
